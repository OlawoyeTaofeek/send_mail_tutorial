```python
def read_file(filepath: str):
    with open(filepath, "r", encoding="utf-8") as f:
        output = f.read()
    return output   
```

```python
read_file("data/Birthday_Wish_Template.txt")
```

```python
def modify(output: str, name: str, your_name: str):
    new_output = output.replace("{name}", name).replace("[Your Name]", your_name)

    return new_output  
```

```python
print(modify(read_file("data/Birthday_Wish_Template.txt"), "Inioluwa", "Taofeek"))
```

```python
import smtplib  
import os

with smtplib.SMTP("smtp.gmail.com", 587) as connection:
    connection.starttls()
    connection.login(user=os.getenv("my_email"), password=os.getenv("my_password"))
    connection.sendmail(
        from_addr=os.getenv("my_email"), 
        to_addrs="ayetoroinioluwa@gmail.com",
        msg=modify(read_file("data/Birthday_Wish_Template.txt"), "Inioluwa", "Taofeek").encode("utf-8")
    )
```
